```python
import numpy as np
```

```python
with open('0_0.txt') as f:
    data_list=[]
    for row in f.readlines():
        print(row)
        data_list.append([x for x in row.replace('\n','')])
data=np.array(data_list,dtype='float')
```

```python
u,s,v=np.linalg.svd(data)
nvec=6
sig=np.diag(s[:nvec])
recon=u[:,:nvec].dot(sig).dot(v[:nvec,:])
np.set_printoptions(threshold=np.nan)
recon[recon>0.8]=1
recon[recon<0.8]=0
recon=recon.astype(np.int)
for i in recon:
    s=''
    for j in i:
        s+=str(j)
    print(s)
    
```
